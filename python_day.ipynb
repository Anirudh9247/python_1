```python
import random
p1=input("player 1 ")
p2=input("player 2 ")
s1=10
s2=10
d1=random.randint(1,10)
d2=random.randint(1,10)
print("_________player 1 turn_________")
while True:


  g2=input("enter your guess")
  s2=s2-1
  if g2==d2:
    break
print("_________Subway_________")
print("{}:{}".format(p1,s1))
print("{}:{}".format(p2,s2))
print("________________________")
if s1>s2:
  print("Winner is",p1)
elif(s2>s1):
  print("Winner is",p2)
else:
  print("DRAW")
```

```python
legs=int(input("Enter num of legs: "))
heads=int(input("Enter total of heads: "))
flag=False
for cows in range(0,heads+1):
  hens=heads-cows
  cal_legs=cows*4+hens*2

  if cal_legs==legs:
    print("Number  of cows: ",cows)
    print("Number of hens: ",hens)
    flag=True
    break
  if flag==False:
    print("No solution")
```

```python
import time
import sys
h=3
m=40
s=0
print("\033[1;31m]")
while True:
  sys.stdout.write("\r{:02d}:{:02d}:{:02d}".format(h,m,s))
  sys.stdout.flush()
  time.sleep(1)
  s=s+1
  if s==60:
    s=0
```
