```python
name1 = input("Enter the first name: ").lower().replace(" ", "")
name2 = input("Enter the second name: ").lower().replace(" ", "")


list_name1 = list(name1)
list_name2 = list(name2)

for char in list_name1:
    if char in list_name2:
        list_name2.remove(char)
        list_name1.remove(char)
remaining_chars = len(list_name1) + len(list_name2)

flames_list = ["F", "L", "A", "M", "E", "S"]

while len(flames_list) > 1:
    split_index = (remaining_chars % len(flames_list)) - 1

    if split_index >= 0:
        right = flames_list[split_index + 1:]
        left = flames_list[:split_index]
        flames_list = right + left
    else:
        flames_list = flames_list[:len(flames_list) - 1]


result = {
    "F": "Friends",
    "L": "Lovers",
    "A": "Affectionate",
    "M": "Marriage",
    "E": "Enemies",
    "S": "Siblings"
}

print(f"The relationship between {name1.capitalize()} and {name2.capitalize()} is: {result[flames_list[0]]}")
```
